### c. Bu kod bloğunda, top250 DataFrame’in 'genre' sütununu kullanarak her bir benzersiz film türünün kaç kere geçtiğini sayıp, Seaborn kütüphanesi ile en çok film yapılan 10 türü bar grafiğiyle gösteriyoruz.

1. ```top_genres = top250["genre"].str.split(",", expand=True).stack().value_counts().largest(10)```

   - Bu satır, top_genres değişkenine en yüksek sayıya sahip olan 10 türü atar.

2. ```sns.set_style("whitegrid")```

   - sns (Seaborn kütüphanesi) ile grafik stilini beyaz ızgaralı olarak ayarlar.

<br />

3. ```plt.figure(figsize=(10,6))```

   - Grafiğin boyutunu belirler: genişlik 10 birim ve yükseklik 6 birim.

<br />

4. ```ax = sns.barplot(x=top_genres.values, y=top_genres.index, hue=top_genres.index, legend=False, palette="colorblind")```

   - ```sns.barplot()```: Seaborn kütüphanesinin bir fonksiyonudur ve bar grafiği çizer.

   - ```x=top_genres.values```: x eksenine film sayısını yerleştirir.
   - ```y=top_genres.index```: y eksenine film türlerini yerleştirir.
   - ```hue=top_genres.index```: y eksenindeki değerlere renklendirme yapar.
   - ```legend=False```: Grafikte Legend gösterilmez.
   - ```palette="colorblind"```: Grafiğin renk paletini belirler.

<br />

5. 
   ```
   for container in ax.containers:
      ax.bar_label(container, fmt='%d', label_type='edge', fontsize=12, color='black', padding=3)
   ```

   - Bu kod satırında, barların yanına y eksenindeki değerlerin, x eksenindeki karşılıklarını yazar.

<br />

6. ```ax.set(xlabel='Number of Movies', ylabel='Genres')```

   - x ekseninin etiketini 'Number of Movies' olarak, y ekseninin etiketini 'Genres' olarak ayarlar.

<br />

7. ```plt.title("Top 10 Movie Genres")```

   - Grafiğin başlığını 'Top 10 Movie Genres' olarak ayarlar.

<br />

8. ```plt.show()```

   - Oluşturulan grafiği gösterir.

````python
# En çok film yapılan ilk 10 türü seç ve top_genres değişkenine ata

top_genres = top250["genre"].str.split(",", expand=True).stack().value_counts().nlargest(10)
top_genres
````

````python
# Bar grafiği oluştur
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))

ax = sns.barplot(   x=top_genres.values, 
                    y=top_genres.index, 
                    hue=top_genres.index, 
                    legend=False, 
                    palette="colorblind") # palette="Blues_d" aşağıdaki grafik renkleri için

for container in ax.containers:
    ax.bar_label(   container, 
                    fmt='%d', 
                    label_type='edge', 
                    fontsize=12, 
                    color='black', 
                    padding=3)

ax.set(xlabel='Number of Movies', ylabel='Genres')

plt.title("Top 10 Movie Genres")

plt.show()
````

### b. Bu kod bloğu, collections kütüphanesi ile birden fazla filmde rol alan oyuncuların rol aldıkları film sayılarına göre ilk 20 oyuncuyu bulur.

**NOTE :** Bu soru için null değerlerin silindiği data seti kullanılacak.

<br />

1. ```from collections import Counter```
    
    - ```Counter``` sınıfı, collections modülünden içe aktarılır.
    
    - Counter, bir hashable nesnenin kaç kez göründüğünü saymak için kullanılır, yani bir frekans sayacıdır.

<br />

2. ```actors_counter = Counter()```

    - Burada, actors_counter adında boş bir Counter nesnesi oluşturulur.

<br />

3. 
    ```
    for actors in movies['actors']:
        actors_list = [actor.strip() for actor in actors.split(',')]
        actors_counter.update(actors_list)
    ```

    - Bu döngü, movies DataFrame'inin ```actors``` sütunundaki her bir satıra (film) uygulanır.

    - Her satırdaki oyuncu isimleri virgülle ayrılır, ardından her bir isim şeritlerinden ```strip()``` temizlenir ve bir liste konur.

    - Daha sonra, ```actors_counter``` nesnesi bu liste ile güncellenir, yani her bir oyuncunun görünme sayısı artar.

<br />

4. ```top_actors = actors_counter.most_common(20)```

    - most_common metodu, actors_counter içindeki elemanları en yaygın olandan en az yaygın olana doğru sıralar ve ilk 20 tanesini döndürür.

    - top_actors bir liste olur ve her elemanı bir tuple'dır: (oyuncu_adı, film_sayısı)

<br />

5. 
    ```top_actors```

    - En çok filmde rol alan ilk 20 oyuncuyu ve onların rol aldıkları film sayılarını gösterir.

````python
df.actors
````

````python
from collections import Counter

actors_counter = Counter()

for actors in df['actors']:
    actors_list = [actor.strip() for actor in actors.split(',')]
    actors_counter.update(actors_list)

top_actors = actors_counter.most_common(20)

top_actors
````
